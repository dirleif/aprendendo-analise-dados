### Exemplo 1

```python
# importando bibliotecas
import seaborn as sns
import matplotlib.pyplot as plt
```

Efetuando leitura de dados e carregando na variável <i>iris</i>. 

A função <i>load_dataset()</i> pertencente a biblioteca Seaborn efetua a leitura de dados e os carrega em um dataset (conjunto de dados). 

https://seaborn.pydata.org/generated/seaborn.load_dataset.html

Em nosso exemplo foi feita a leitura acerca da Iris, uma flor. Tais dados são lidos a partir da biblioteca <i>statsmodels</i> e/ou <i>scikit-learn</i>. São dados que são instalados/baixados com instalação do Python e de suas bibliotecas. Os dados são carregados do arquivo que contêm as informações da flor Iris para o variável <i>iris</i>. Existem muito outros datasets que o Python já instala por padrão.

```python
# carrega dataset na variável
iris = sns.load_dataset('iris')

# solicitada a impressão dos dados com gráfico estilo swarm plot
sns.swarmplot(x='species', y='petal_length', data=iris)

# imnpressão gráfico
plt.show()
```

No eixo X teremos as espécies de Iris, no caso temos 3 espécies (setosa, versicolor species e virginica). No eixo Y o tamanho da pétala. Pelo gráfico gerado é claro que a espécie setosa é a que possui as menores pétalas enquanto que a espécie virginica são as que possuem as maiores pétalas.

<i>Setar</i> o programa para considerar as configuração padrão do <b>matplotlib</b>.
```python
# importando bibliotecas
import matplotlib.pyplot as plt
plt.style.use('classic')
import numpy as np
import pandas as pd
```

A diferença entre o exemplo 1 é a de que os <i>labels</i> do eixo X e Y foram alterados e além disso foram informadas as coordenadas de X e de Y, formando os pares ordenados (1,1), (2,8), (3,27), (4, 64).

```python
# setar variável para gerar valor aleatório, randômico.
rng = np.random.RandomState(0)
```

Criada a variável <i>x</i> (lista) onde o primeiro elemento da lista será igual a 0, o último elemento igual a 10 e ao todo teremos 100 elementos espaçados de forma regular, ou seja, a lista terá como elementos os valores: 0, 0.101, 0.202, 0.303 e assim sucessivamente até que se atinja o valor 10.

```python
x = np.linspace(0, 10, 100)
```

A variável <i>y</i> será composta de 5 listas de 100 elementos cada. Os elementos serão gerados de forma aleatória <b>(randn)</b>. Cada elemento de <i>y</i> será igual a soma de todos os elementos anteriormente gerados - relacionado a lista ao qual pertence - mais o elemento que foi gerado na etapa corrente de execução - função <b>cumsum</b>. Esta lógica foi adotada para que os 5 grupos gerados não ficassem sobrepostos uns sobre os outros mas para que tivessem certa independência/divergência.

```python
y = np.cumsum(rng.randn(100, 5), 0)
```

```python
# gera gráfico
plt.plot(x, y)

# gera a legenda na parte superior esquerda
plt.legend('ABCDE', ncol=2, loc='upper left');
```

```python
# importando biblioteca
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# setar para usar Seaborn
sns.set()

# definindo variáveis
grupos = 5
anterior = (30, 20, 18, 17, 15)
atual = (40, 20, 20, 10, 10)

# criação dos grupos
index = np.arange(grupos)
# largura das barras
bar_width = 0.3

# barras vermelhas
barras1 = plt.bar(index, anterior,bar_width,color='r',label='10-10-2018')
# barras azuis 
barras2 = plt.bar(index + bar_width, atual, bar_width,color='b',label='17-10-2018')

# labels
plt.xlabel('Candidato')
plt.ylabel('Percentual')
# título
plt.title('Pesquisa Eleitoral')
plt.xticks(index + bar_width, ('Leonardo', 'Maria', 'Ana', 'Outros', 'Indecisos'))
# mostra legendas e gráfico
plt.legend()
plt.show()
```

### Exemplo 6

```python
# importando biblioteca
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# setar para usar a configuração padrão do matplotlib
plt.style.use('classic')

# criação da variável rng para geração de números aleatórios/randômicos
rng = np.random.RandomState(5)
```

Geração de 300 valores no eixo <i>X</i>, com valores variando entre 0 e 10, espaçados de forma regular e ordenada.

Geração de 300 valores aleatórios, somente 1 grupo de valores. Valores serão armazenados na variável <i>Y</i>.

Geração de 300 valores aleatórios, somente 1 grupo de valores. Valores serão armazenados na variável <i>Z</i>.

```python
X = np.linspace(0, 10, 300)
Y = rng.rand(300, 1)
Z = rng.rand(300, 1)
```

```python
# configurar para gráfico a ser gerado, ser gerado em 3D
ax = plt.figure().add_subplot(111,projection='3d')
 
# geração dos pontos com coordenadas X, Y e Z, na cor vermelha (r - red)
ax.scatter(X,Y,Z, color='r')
# configurar labels dos eixos X, Y e Z
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_zlabel('Eixo Z')
# gerar gráfico 
plt.show()
```

### Exemplo 7

O exemplo a seguir foi retirado de https://python-graph-gallery.com/371-surface-plot/

```python
# libraries
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
 
# Get the data (csv file is hosted on the web)
url = 'https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/volcano.csv'
data = pd.read_csv(url)
 
# Transform it to a long format
df=data.unstack().reset_index()
df.columns=["X","Y","Z"]
 
# And transform the old column name in something numeric
df['X']=pd.Categorical(df['X'])
df['X']=df['X'].cat.codes
 
# Make the plot
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_trisurf(df['Y'], df['X'], df['Z'], cmap=plt.cm.viridis, linewidth=0.2)
plt.show()
 
# to Add a color bar which maps values to colors.
fig = plt.figure()
ax = fig.gca(projection='3d')
surf=ax.plot_trisurf(df['Y'], df['X'], df['Z'], cmap=plt.cm.viridis, linewidth=0.2)
fig.colorbar( surf, shrink=0.5, aspect=5)
plt.show()
 
# Rotate it
fig = plt.figure()
ax = fig.gca(projection='3d')
surf=ax.plot_trisurf(df['Y'], df['X'], df['Z'], cmap=plt.cm.viridis, linewidth=0.2)
ax.view_init(30, 45)
plt.show()
 
# Other palette
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_trisurf(df['Y'], df['X'], df['Z'], cmap=plt.cm.jet, linewidth=0.01)
plt.show()
```
