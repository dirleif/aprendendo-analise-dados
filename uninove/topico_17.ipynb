### Histograma

```python
# importando bibliotecas
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
```

Na variável <i>x</i> temos os valores que serão analisados. Note que a amostra é bem pequena, somente a título de ilustração. O menor valor é 1 e o maior valor 80, ou seja, temos um intervalo de 80 valores.

<i>num_bins</i> é definido quantos intervalos desejamos criar. O programa irá criar 4 intervalos regulares em se considerando que o intervalo completo inicia em 1 e vai até 80. Os 4 intervalos criados serão: 1-20, 21-40, 41-60 e 61-80.

A seguir o gráfico é impresso informando quantos números pertencentes a lista estão em cada intervalo. No interlalo de 41-60 temos 0 ocorrências, por outro lado no intervalo de 1-20 temos 10 ocorrências (1,4,7,8,10,20,14, 5, 12 e 9).

```python
x = [1,4,7,8,10,20,25,28,14,32,35,38,5,65,77,80,12,9]

num_bins = 4

plt.hist(x, num_bins, facecolor='green')
plt.show()
```

### Pizza

```python
# importando biblioteca
import matplotlib.pyplot as plt

labels = 'Pretos', 'Castanhos', 'Verdes', 'Azuis'
sizes = [70, 80, 20, 10]

# impressão gráfico
# parâmetro autopct é usado para mostrar o percentual de cada opção
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.axis('equal')  
plt.show()
```

### Exemplo 11

```python
# importando biblioteca
import matplotlib.pyplot as plt

labels = 'Pretos', 'Castanhos', 'Verdes', 'Azuis'
sizes = [70, 80, 20, 10]

# espaçamento entre os dados
explode = (0.03, 0.03, 0.03, 0.03)

# shadow - mostrar sombra sobre a fatia
# startangle - ângulo em que o gráfico inicia sua impressão
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')  
plt.show()
```

### Barras

```python
# importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt

# definido grupos
grupos = 5

# percentual de cada candidato
anterior = (30, 20, 18, 17, 15)
atual = (40, 20, 20, 10, 10)

# criação dos 5 grupos
index = np.arange(grupos)

# largura de cada barra
bar_width = 0.3

# barra vermelha
barras1 = plt.bar(index, anterior,bar_width,color='r',label='10-10-2018')

# barra azul - observe o deslocamento
barras2 = plt.bar(index + bar_width, atual, bar_width,color='b',label='17-10-2018')
 
plt.xlabel('Candidato')
plt.ylabel('Percentual')
plt.title('Pesquisa Eleitoral')
plt.xticks(index + bar_width, ('Leonardo', 'Maria', 'Ana', 'Outros', 'Indecisos'))

# apresenta legendas
plt.legend()
plt.show()
```

```python
# importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt

# definindo grupos
grupos = 5

# percentual de cada candidato
anterior = (30, 20, 18, 17, 15)
atual = (40, 20, 20, 10, 10)

# criação dos 5 grupos
index = np.arange(grupos)

# largura de cada barra
bar_width = 0.3

# barra vermelha
barras1 = plt.barh(index, anterior,bar_width,color='r',label='10-10-2018')

# barra azul - observe o deslocamento
barras2 = plt.barh(index + bar_width, atual, bar_width,color='b',label='17-10-2018')
 
plt.ylabel('Candidato')
plt.xlabel('Percentual')
plt.title('Pesquisa Eleitoral')
plt.yticks(index + bar_width, ('Leonardo', 'Maria', 'Ana', 'Outros', 'Indecisos'))

# apresenta legendas
plt.legend()
plt.show()
```
